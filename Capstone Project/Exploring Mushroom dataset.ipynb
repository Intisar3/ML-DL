```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
mushroom = fetch_ucirepo(id=73) 
  
# data (as pandas dataframes) 
X = mushroom.data.features 
y = mushroom.data.targets 
  
# metadata 
print(mushroom.metadata) 
  
# variable information 
print(mushroom.variables) 
```

```python
# Combine features and target into one DataFrame
df = pd.concat([X, y], axis=1)
```

```python
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='poisonous', palette='Set2')
plt.title('Distribution of Edible vs. Poisonous Mushrooms')
plt.xlabel('Mushroom Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='cap-color', hue='poisonous', palette='Set1')
plt.title('Cap Color vs Poisonous')
plt.xlabel('Cap Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='odor', hue='poisonous', palette='Set3')
plt.title('Odor vs Poisonous')
plt.xlabel('Odor Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='gill-size', hue='poisonous', palette='coolwarm')
plt.title('Gill Size vs Poisonous')
plt.xlabel('Gill Size')
plt.ylabel('Count')
plt.show()
```

```python
crosstab = pd.crosstab(df['population'], df['habitat'])

plt.figure(figsize=(10, 6))
sns.heatmap(crosstab, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Population vs Habitat Count')
plt.xlabel('Habitat')
plt.ylabel('Population')
plt.show()
```

```python
important_features = ['cap-shape', 'gill-color', 'ring-type', 'spore-print-color']
for col in important_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, hue='poisonous')
    plt.title(f'{col} vs Poisonous')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
