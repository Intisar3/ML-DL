```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yyhat = np.linspace(-2, 2, 400)
hinge_loss = np.maximum(0, 1 - yyhat)
```

```python
plt.plot(yyhat, hinge_loss, label='Hinge Loss')
plt.axvline(x=1, color='green', linestyle='--', label='yŷ = 1 (zero loss threshold)')
plt.xlabel('yŷ = y * ŷ')
plt.ylabel('Hinge Loss')
plt.title('Hinge Loss vs yŷ')
plt.legend()
plt.grid(True)
plt.show()
```
